```python
#Hacemos la petición a la base de datos del gobierno
client = Socrata("www.datos.gov.co", None)
results = client.get("f789-7hwg", where="anno_cargue_secop >= '2018' AND departamento_entidad = 'Nariño' AND estado_del_proceso = 'Celebrado' AND tipo_de_contrato != 'Prestación de Servicios'", limit=50000)

#Lo convertimos a un Data Frame
pd.options.display.max_columns = None #Muestra todas las columnas completas
results_df = pd.DataFrame.from_records(results)

#Filtrar solo las columnas que necesitamos
columnas_a_mantener = ['anno_firma_contrato', 'nombre_entidad', 'municipio_entidad', 'detalle_del_objeto_a_contratar', 'tipo_de_contrato', 'cuantia_contrato', 'numero_de_constancia']
df_reducido = results_df[columnas_a_mantener]
```
